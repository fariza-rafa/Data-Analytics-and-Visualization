# **Exploring data sheet of Samsung R&D Institute Bangladesh Ltd.**




```python
s = pd.read_csv('employee_data_20.csv')
print(s.head)
```

# **Vizualising the salary of the employees in Bar Chart**

```python
fig, ax = plt.subplots()
ax.barh(s['Name'],s['Salary'])
ax.set_title('Salary of Employees')
ax.set_xlabel('Amount of Salary')
ax.set_ylabel('Employees')
ax.legend()
plt.show()
```

# **Vizualising the salary of various departments in Bar Chart**

```python
fig, ax = plt.subplots()
ax.barh(s['Department'],s['Salary'])
ax.set_title('Salary of Employees')
ax.set_xlabel('Amount of Salary')
ax.set_ylabel('Departments')
ax.legend()
plt.show()
```

# **Vizualising the total salary of various departments in Bar Chart**

```python
d = s.groupby('Department').sum()
d = d.reset_index()

fig, ax = plt.subplots()
ax.barh(d['Department'],d['Salary'])
ax.set_title('Salary of Employees')
ax.set_xlabel('Amount of Salary')
ax.set_ylabel('Departments')
ax.legend()
plt.show()
```

# **Vizualising various position's salary of employees in Bar chart**

```python
g = s.groupby('Position').sum()
g = g.reset_index()

fig, ax = plt.subplots()
ax.barh(g['Position'],g['Salary'])
ax.set_title('Salary of Employees')
ax.set_xlabel('Amount of Salary')
ax.set_ylabel('Departments')
ax.legend()
plt.show()
```

# **Vizualising the location of employees in Bar chart**

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.barh(s['Location'],s['Name'])
ax.set_title('Location of Employees')
ax.set_xlabel('Name')
ax.set_ylabel('Location')
ax.legend()
plt.show()
```

# **Vizualising the genders in Pie chart**

```python
g = s.groupby('Gender').sum()
g = g.reset_index()

fig, ax = plt.subplots()
ax.pie(g['Salary'])
ax.set_title('Gender of Employees')

ax.legend(g['Gender'])
plt.show()
```

# **Vizualising various departments of SRBD in slices with colors in Pie chart**

```python
slices = [40,20,60,15,15]
depts = ['HR','IT','ENGINEERING','MARKETING','FINANCE']
cols = ['Cyan','Blue','Red','Yellow','Brown']

plt.pie(slices,labels=depts,colors=cols,startangle=90,shadow=True,explode=(0.1,0,0.1,0,0),autopct='%1.2f')

plt.title('Departments at SRBD')
plt.legend()
plt.show()
```
